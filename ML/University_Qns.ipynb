#1. What is the equation of least squares ?


The equation of least squares represents the best-fitting linear relationship between two variables, typically denoted as y (the dependent variable) and x (the independent variable). It is often used in regression analysis to find a linear equation that minimizes the sum of the squared differences between the observed values (the data points) and the values predicted by the equation. The equation of least squares is typically in the form of:

y = mx + b

Where:

y is the dependent variable (the variable you want to predict or explain).
x is the independent variable (the variable used to make predictions).
m is the slope of the line, representing how much y changes for a unit change in x.
b is the y-intercept, which is the value of y when x is 0.

There are various Performance matrix in classification model

1. Accuracy
2. TPR (True Positive Rate)
3. TNR (True Negative Rate)
4. FPR (Flase Positive Rate)
5. FNR (False Negative Rate)
6. Precision
7. Recall
8. F1 score



1. Accuracy = (TP+TN) / (TP+TN+FP+FN)

```python
accuracy = 150 / 165
print(accuracy)
```

2. True Positive Rate

TPR = TP / Actual Positive = TP / (TP + FN)

```python
TPR = 100 / 105
print(TPR)

# Ideally TPR value should be as high as possible
# TPR Range (0-1)
```

3. True Negative Rate

TNR = TN / Actual Negative = TN / (TN + FP)

```python
TNR = 50 / 60
print(TNR)

# Ideally TNR value should be as high as possible
# TNR Range (0-1)
```

4. False Positive Rate

FPR = FP / Actual Positive = FP / (FN + TP)

```python
FPR = 10 / 105
print(FPR)
# Ideally FPR value should be as low as possible
# TNR Range (0-1)
```

5. False Negative Rate

FNR = FN / Actual Negative = FN / (TN + FP)


```python
FNR = 5/ 60
print(FNR)

# Ideally FNR value should be as low as possible
# TNR Range (0-1)
```

6. Precision - Out of the total positive predicted by the model what is the percentage of actual positive results

Precision = TP / Predicted Positive = TP / (TP + FP)

```python
Precision = 100 / 110
print(Precision)

# Ideally Precision value should be as high as possible
# Precision Range (0-1)
```

7. Recall - Simillar to True Positive Rate

Recall = TP / Actual Positive = TP / (TP+FN)

```python
Recall = 100 / 105
print(Recall)

# Ideally Recall value should be as high as possible
# Recall Range (0-1)
```

8. F1 Score

F1 Score = 2 * ( (Precision * Recall ) / (Precision + Recall ) )

```python
x = Precision * Recall
y = Precision + Recall

z = x/y
f1_score = 2*z
print(f1_score)


# Ideally F1 Score value should be as high as possible
# F1 Score Range (0-1)
```
