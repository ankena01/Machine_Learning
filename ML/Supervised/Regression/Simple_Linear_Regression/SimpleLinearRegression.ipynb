# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

# Read the dataset

```python
df = pd.read_csv("/content/drive/MyDrive/IBMR/ML/Supervised/Regression/Simple_Linear_Regression/homeprices.csv")
```

```python
df
```

# Visualizing the data

```python
plt.xlabel("Area (sq ft)")
plt.ylabel("Price ($)")
plt.scatter(df['Area'] , df['Price'], marker='*', color = 'red')
```

# Create a Linear Regression Model

```python
linear_reg = LinearRegression()
```

#Train the model

```python
linear_reg.fit(df[['Area']], df['Price'])
```

```python
linear_reg
```

# Make Predictions (Test the model)

```python
linear_reg.predict([[3300]])
```

#finding slope and y-intercept

```python
# slope

linear_reg.coef_
```

```python
# y-intercept

linear_reg.intercept_
```
