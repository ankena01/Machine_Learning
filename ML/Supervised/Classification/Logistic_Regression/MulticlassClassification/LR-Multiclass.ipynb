#Split the dataset in training and test data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

# Create a Logisctic Regression Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

# Train the model

```python
model.fit(X_train, y_train)
```

# Test the model


```python
model.predict(X_test)
```

```python
# Make prediction for first 5 samples

model.predict(digits.data[0:5])
```

# Model Accuracy

```python
model.score(X_test, y_test)
```

#Explore Model Performance using Confusion Matrix

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

# Visualise Confusion Matrix using Seaborn


```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
