# Create a Dataframe

```python
df = pd.DataFrame(data=digits['data'], columns=digits['feature_names'])
```

```python
df.head()
```

Add target label to the dataframe

```python
df['target'] = digits['target']
```

```python
df.columns
```

```python
df.head()
```

#Dividing the dataset into Independent and Target Label

```python
X = df.drop('target', axis = 1)
X.columns
```

```python
y = df['target']
y
```

# Splitting the dataset in Train and Test Set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

#Create model using RandomForest Classifier

ensemble means multiple algorithms. Since we are using multiple decision trees to classify the outcome the randomforest classifier is appropriately palced under ensemble class of sklearn library

```python
from sklearn.ensemble import RandomForestClassifier
```

Create the model

Default 100 decision tree will be created (n_estimators=100)

```python
model = RandomForestClassifier(n_estimators=100)
```

#Train the model

```python
model.fit(X_train,y_train)
```

```python
help(model)
```

#Test the model

```python
model.predict(X_test)
```

# Accuracy of  the model

```python
model.score(X_test, y_test)
```

#Evaluate model performance using confusion matrix

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

Visualise the confusion matrix using seaborn and matplotlib

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot = True)
plt.xlabel('Predict')
plt.ylabel('Truth')
```
